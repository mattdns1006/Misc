```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def func(x1):
    return 1.2*((x1-2)**2)

def dfunc(x1):
    return 2.4*(x1-2)
```

```python
xt = None
x = np.arange(-30,30,0.01)
y = func(x)
lr = 0.01

for t in range(100):
    
    if t ==0:
        xt = np.random.uniform(-30,30) #Initial guess
    
    fxt = func(xt)
    
    if t % 10 == 0:
        plt.plot(x,y)
        plt.title("step %d with lr = %f current value %f and gradient %f" % (t,lr,fxt,dfxt))
        plt.plot(xt,fxt,"ro")
        plt.plot(2,func(2),"bo")
        plt.show()

    dfxt = dfunc(xt)
    xt -= lr*dfxt
    
    
```

```python
def rosenbrock(x1,x2,b):
    return (1-x1**2)  + b*((x2-x1**2)**2)

def dRosenbrock(x1,x2,b):
    pdRosenbrockx1 = -2 + 2*x1 - 2*b*x1*(x2-x1**2)
    pdRosenbrockx2 = 2*b*(x2-x1**2)
    return np.array([pdRosenbrockx1, pdRosenbrockx2])
```
