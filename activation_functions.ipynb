```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import expit as sigmoid
plt.style.use('ggplot')
save_path = "/Users/matt/University/writeUps/Mres/images/method/"
```

```python
x = np.linspace(-5,5,1000)
relu = lambda x: x if x>0 else 0
relu_prime = lambda x: 1 if x>0 else 0
relu_prime = np.vectorize(relu_prime,otypes=[np.float32])
relu = np.vectorize(relu,otypes=[np.float32])
```

```python
y1 = sigmoid(x)
y1_prime = sigmoid(x)*(1-sigmoid(x))
y2 = np.tanh(x)
y2_prime = 1-np.tanh(x)**2
y3 = relu(x)
y3_prime = relu_prime(x)
```

```python
f, (ax1,ax2,ax3) = plt.subplots(ncols=3,figsize=(20,6))
save_path
ax1.plot(x,y1)
ax1.plot(x,y1_prime)
ax1.set_title('Sigmoid')

ax2.plot(x,y2)
ax2.plot(x,y2_prime)
ax2.set_title('Tanh')

ax3.plot(x,y3,label='f(x)')
ax3.plot(x,y3_prime,label="f'(x)")
ax3.set_title('ReLU')

for ax in [ax1,ax2,ax3]:
    ax.set_xlabel("x",fontsize=20);ax.set_ylabel("y",fontsize=20)
    ax.set_ylim(bottom=-1.1,top=1.5)
handles, labels = ax3.get_legend_handles_labels()
plt.figlegend( handles,labels=('f(x)',"f'(x)"),
              loc = 3, ncol=2, labelspacing=0.5,
              fontsize=15, bbox_to_anchor=(0.04,-0.01))
plt.savefig(save_path+"af.png",bbox_inches='tight')
plt.show()
```
