```python
import numpy as np
import numpy.random as rng
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
import pylab
```

```python
sample = rng.exponential(3,10000)
x = np.linspace(-15,15,1000)
dNorm = stats.norm.pdf(x,0,3)
plt.figure(figsize=(35,10))
plt.subplot(221)
plt.hist(sample,normed=1,bins=100,alpha=0.4,color="red")
#plt.plot(x,dNorm)

def boxCox(x,l = 1.5):
    x = x.astype(np.float32)
    return (x**l - 1)/ l
    
sampleX = boxCox(sample,0.3)
plt.subplot(222)
plt.hist(sampleX,normed=1,bins = 100,alpha=0.4,color="green")
plt.subplot(223)
stats.probplot(sample, dist="norm", plot=pylab)
plt.subplot(224)
stats.probplot(sampleX, dist="norm", plot=pylab)
plt.show()
```
